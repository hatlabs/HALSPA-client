```python
def _reverse_bits(value, num_bits):
    """_summary_
    Reverse the order of bits in a given integer value.

    The lowest bit of the value will become the highest bit, and so on.

    Args:
        value (int): The integer value whose bits are to be reversed.
        num_bits (int): The number of bits in the integer value.
    """
    reversed_value = 0
    for i in range(num_bits):
        if value & (1 << i):
            reversed_value |= 1 << (num_bits - 1 - i)
    return reversed_value
```

```python
from contextlib import contextmanager

@contextmanager
def resource():
    print("Enter")
    yield "value"
    print("Exit")

def get_resource(use_context=False):
    if use_context:
        return resource()
    else:
        # Simulate one-shot usage
        with resource() as r:
            return r

# Regular function
val = get_resource()
print(f"Function: {val}")

# Context manager
with get_resource(use_context=True) as v:
    print(f"Context: {v}")
```
